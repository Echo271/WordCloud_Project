First step, import all the modules and libraries that the project needs. Such as Wordcloud, IO, Sys, Matplotlib, and Display

```python
import wordcloud
import io
import sys
from matplotlib import pyplot as plt
from IPython.display import display
```

And then we also need the nltk module to get stopwords

```python
import nltk
from nltk.corpus import stopwords

nltk.download("stopwords")
```

Create a function to process paragraphs start to split and remove punctuation and stopwords to be list, and then we count each word into a dictionary called freq. And then we put freq dictionary into wordcloud function

```python
def calculate_frequencies(file_contents):
    # Here is a list of punctuations and uninteresting words you can use to process your text
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    stop_words = stopwords.words("english")

    # LEARNER CODE START HERE
    freq = {}
    
    for l in punctuations:
        file_contents = file_contents.replace(l," ") #remove punctuations from file_contents
        
    for i in stop_words:
        i = " "+i+" "
        file_contents = file_contents.lower().replace(i," ") #remove unisteresting_words
        
    filecontent_list = file_contents.split()

#print(calculate_frequencies(file_contents))
    for word in filecontent_list:
        if word not in freq:
            freq[word] = 1
        else:
            freq[word] += 1
                
    
    #wordcloud
    cloud = wordcloud.WordCloud()
    cloud.generate_from_frequencies(freq)
    return cloud.to_array()
```

After the script above is run, drop the paragraphs or sentences will you make it to be  a wordcloud here

```python
file_contents = input("Drop your text here: ")
```

Finally, you have finished the process and get the wordcloud images 

```python
# Display your wordcloud image
myimage = calculate_frequencies(file_contents)
plt.figure(figsize = (10,5))
plt.imshow(myimage, interpolation = 'nearest', cmap = 'Dark2')
plt.axis('off')
plt.show()
```
